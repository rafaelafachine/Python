Exemplo - Avaliação de Veículos

```python
import pandas as pd
import random
import numpy as np
```

```python
# Lista de exemplos de atributos dos carros

marcas = ["Toyota", "Ford", "Chevrolet", "Honda", "Hyundai", "BMW", "Mercedes", "Volkswagen", "Audi", "Nissan"]
anos = list(range(2000, 2023))  # Anos de 2000 a 2022
quilometragens = [random.randint(0, 200000) for _ in range(500)]  # Quilometragem aleatória entre 0 e 200.000 km
seguro = ["Sim", "Não"]

#Criando um dataframe
df = pd.DataFrame({
    # random.choices 'sorteiam' números aleatórios
    "Marca": random.choices(marcas, k=500),
    "Ano de Fabricação": random.choices(anos, k=500),
    "Quilometragem": quilometragens,
    "Seguro": random.choices(seguro, k=500)
})

# Mostrando as primeiras 5 linhas do dataframe
df.head()
```

```python
# Ano de Fabricação vs Quilometragem

from matplotlib import pyplot as plt
df.plot(kind='scatter', x='Ano de Fabricação', y='Quilometragem', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```
